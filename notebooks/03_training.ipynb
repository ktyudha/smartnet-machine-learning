# 03 — Training & Evaluation
Training Decision Tree untuk klasifikasi relay LoRa.

```python
import sys
sys.path.append('..')

from src.preprocessing import load_dataset, clean, split_features, scale
from src.training.train import train
from src.training.evaluate import evaluate
from src.training.save_model import save

df = load_dataset()
df = clean(df)
X, y = split_features(df)
X_scaled, scaler = scale(X)

model, X_test, y_test = train(X_scaled, y)
print('Training selesai.')
```

```python
metrics = evaluate(model, X_test, y_test)
print(f"Accuracy : {metrics['accuracy']:.4f}")
print(f"\nClassification Report:\n{metrics['report']}")
```

```python
import matplotlib.pyplot as plt

cm = metrics['confusion_matrix']
labels = ['Direct', 'Relay']

fig, ax = plt.subplots(figsize=(5, 4))
im = ax.imshow(cm, cmap='Blues')
plt.colorbar(im, ax=ax)
ax.set_xticks([0, 1])
ax.set_yticks([0, 1])
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix')
for i in range(2):
    for j in range(2):
        ax.text(j, i, str(cm[i, j]), ha='center', va='center', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import plot_tree

fig, ax = plt.subplots(figsize=(14, 6))
plot_tree(
    model,
    feature_names=['rssi', 'snr'],
    class_names=['Direct', 'Relay'],
    filled=True,
    max_depth=3,
    ax=ax,
)
ax.set_title('Decision Tree (max_depth=3 untuk visualisasi)')
plt.tight_layout()
plt.show()
```

```python
save(model, scaler)
print('Model dan scaler tersimpan di models/')
```

# EDA — Dataset RSSI/SNR
Eksplorasi dataset untuk klasifikasi keputusan relay LoRa.

- **Feature**: `rssi`, `snr`
- **Target**: `relay` (0 = Direct, 1 = Relay)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/dataset_rssi_snr.csv')
df.head()
```

```python
print(df.shape)
print(df.dtypes)
df.describe()
```

```python
print('Missing values:')
print(df.isnull().sum())
print(f'\nDuplicate rows: {df.duplicated().sum()}')
```

## Class Balance

```python
counts = df['relay'].value_counts()
labels = ['Direct (0)', 'Relay (1)']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(labels, counts.values, color=['steelblue', 'tomato'])
axes[0].set_title('Class Distribution')
axes[0].set_ylabel('Count')
for i, v in enumerate(counts.values):
    axes[0].text(i, v + 20, str(v), ha='center')

axes[1].pie(counts.values, labels=labels, autopct='%1.1f%%', colors=['steelblue', 'tomato'])
axes[1].set_title('Class Proportion')

plt.tight_layout()
plt.show()
```

## Distribusi Feature per Kelas

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

for label, color in [(0, 'steelblue'), (1, 'tomato')]:
    subset = df[df['relay'] == label]
    axes[0].hist(subset['rssi'], bins=30, alpha=0.6, color=color, label=f'relay={label}')
    axes[1].hist(subset['snr'],  bins=30, alpha=0.6, color=color, label=f'relay={label}')

axes[0].set_title('Distribusi RSSI')
axes[0].set_xlabel('RSSI (dBm)')
axes[0].legend()

axes[1].set_title('Distribusi SNR')
axes[1].set_xlabel('SNR (dB)')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Korelasi Feature → Target

```python
print(df[['rssi', 'snr', 'relay']].corr())
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
corr = df[['rssi', 'snr', 'relay']].corr()
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(corr.columns)))
ax.set_yticks(range(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
for i in range(len(corr)):
    for j in range(len(corr.columns)):
        ax.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center', fontsize=10)
ax.set_title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

## Scatter Plot RSSI vs SNR

```python
fig, ax = plt.subplots(figsize=(8, 5))

for label, color, name in [(0, 'steelblue', 'Direct'), (1, 'tomato', 'Relay')]:
    subset = df[df['relay'] == label]
    ax.scatter(subset['rssi'], subset['snr'], alpha=0.3, s=10, color=color, label=name)

ax.set_xlabel('RSSI (dBm)')
ax.set_ylabel('SNR (dB)')
ax.set_title('RSSI vs SNR per Kelas')
ax.legend()
plt.tight_layout()
plt.show()
```
